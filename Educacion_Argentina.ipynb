```python
# prompt: Armar grafico de barras con Promedio general de cada materia, ordenados de mayor a menor, agregar las faltas como valor constrastable en dos ejes. Agregar las leyendas que ayuden a la interpretación.

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data (replace with your actual data)
data = {
    'Materia': ['Matemáticas', 'Física', 'Química', 'Historia', 'Literatura'],
    'Promedio': [8.5, 7.2, 9.1, 6.8, 8.0],
    'Faltas': [2, 5, 1, 7, 3]
}
df = pd.DataFrame(data)

# Sort by average grade in descending order
df = df.sort_values('Promedio', ascending=False)


# Create the bar chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar plot for average grades
ax1.bar(df['Materia'], df['Promedio'], color='skyblue', label='Promedio')
ax1.set_xlabel('Materia')
ax1.set_ylabel('Promedio', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_ylim(0, 10)  # Set y-axis limit for grades


# Create a second y-axis for absences
ax2 = ax1.twinx()
ax2.plot(df['Materia'], df['Faltas'], marker='o', color='red', linestyle='-', label='Faltas')
ax2.set_ylabel('Faltas', color='red')
ax2.tick_params(axis='y', labelcolor='red')


# Customize the plot
plt.title('Promedio General de Materias (Ordenado de Mayor a Menor) y Faltas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()


# Combine legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')


# Show the plot
plt.show()
```
